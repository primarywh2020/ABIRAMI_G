#### Q. Draw histogram for Age, using Matplotlib

```python
print(titanic.Age)
bins = 5
plt.hist(titanic.Age, bins = bins, edgecolor = 'black',color='pink')
```

#### Q. Make a Pie Chart of Corona Cases by taking numbers list as [500000, 1800000, 1200000] and labels list as ["Deaths", "Total Cases", "Cured"]

```python
#Creating dataset
coronacases = ['Deaths', 'Total Cases', 'Cured']
        
numbers = [500000, 1800000, 1200000]
  
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(numbers, labels = coronacases)
  
# show plot
plt.show()
```
